# Task 1: Store Customer Orders



````python
#list of customers
customers=["Jimmy","Sally","Tom","Jill","Sam","Michael","Sarah","Martha","Deanne","Annie"]

#list of orders
orders=[
("Jimmy", "Television", 400, "Electronics"),
("Sally", "Blouse",75, "Clothing"),
("Tom", "Blender",40, "Home Essentials"),
("Jill", "Headphones", 100, "Electronics"),
("Sam", "Pants", 60, "Clothing"),
("Michael", "Microwave", 50, "Home Essentials"),
("Sarah", "DVD Player", 25, "Electronics"),
("Martha", "Jacket", 120, "Clothing"),
("Deanne", "Toaster", 30, "Home Essentials"),
("Annie", "Monitor", 150,"Electronics")
]

#dictionary of products customers brought

customer_orders={}

for name, product, price, category in orders:
    if name not in customer_orders:
        customer_orders[name]=[]
    customer_orders[name].append((product, price, category))

print("Customers:", customers)
print("Orders:", orders)
print("Customer Orders:", customer_orders)
````

# Task 2: Classify Products by Category


````python
product_categories={}

# Loop through each order and map the product to its category
for name, product, price, category in orders:
    product_categories[product]= category


# Create a set of all unique categories
print("Product Categories:", product_categories)
unique_categories=set(product_categories.values())
print("Unique Categories:", unique_categories)
````

# Task 3: Analyze Customer Orders


````python
#Dictionaty to store total spending per customer

customer_totals={}

#Calculate total amount each customer spends

for customer, items in customer_orders.items():
    total=sum(price for product, price, category in items)
    customer_totals[customer]=total

print("Customer Totals:", customer_totals)

#Customer classification based on spending levels

customer_classification={}

for customer, total in customer_totals.items():
    if total>100:
      customer_classification[customer]="High-value buyer"
    elif 50<= total <=100:
      customer_classification[customer]="Moderate buyer"
    else:
      customer_classification[customer]="Low-value buyer"

print("Customer Classification:", customer_classification)

````

# Task 4: Generate Business Insights






````python
# Total revenue per product category

category_totals={}

for name, product, price, category in orders:
    if category not in category_totals:
        category_totals[category]=0
    category_totals[category]+=price

print("Total Revenues:", category_totals)

#unique products from all orders
unique_products=set(product for name,product, price, category in orders)
print("Unique Products:", unique_products)

#customers who purchased electronincs
electronic_customers=set(name for name, product, price, category in orders if category=="Electronics")
print("Customers who Purchased Electronics:", electronic_customers)

#top three-highest spending officers
top_three=sorted(customer_totals.items(), key=lambda x: x[1], reverse=True)[:3]
print("Top Three-Highest Spending Customers:", top_three)
````

# Task 5: Organize and Display Data





````python
#summary of each customer's total spending and classification
print ("Customer Summary:")
for customer in customer_totals:
  print(customer, "spent$", customer_totals[customer], "-", customer_classification[customer])

#dictionary of categories per customer
customer_categories={}

for name,product, price, category in orders:
  if name not in customer_categories:
   customer_categories[name]=set()
  customer_categories[name].add(category)

  #customers who purchased from more than one category
multi_category_customers={customer for customer, cats in customer_categories.items() if len(cats)>1}
print("Customers Who Purchased From Multiple Categories:", multi_category_customers)

#customers who bought clothing
clothing_buyers= {name for name, product, price, category in orders if category=="Clothing"}

#customers who bought electronics
electronic_buyers= {name for name, product, price, category in orders if category=="Electronics"}

#customers who bought electronics and clothing
both_electronics_and_clothing=electronic_buyers & clothing_buyers

print ("Customers Who Bought Electronics and Clothing:", both_electronics_and_clothing)
````
